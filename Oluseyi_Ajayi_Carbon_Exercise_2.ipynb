# **SAMPLING OF TRAINING DATA TO BALANCE THE DATA**

```python
dff=pd.read_csv('/content/gdrive/MyDrive/Ex2 - Data.csv')
dff
```

```python
dff=dff.drop(columns=['Unnamed: 0','ID'])
```

```python
x_training=dff[df.Fold=='IS']
x_testing=dff[df.Fold=='OS']
y_training=x_training.Target
y_testing = x_testing.Target
```

```python
x_training=x_training.drop(columns=['Fold','Target'])
x_testing=x_testing.drop(columns=['Fold','Target'])
```

```python
x_training
```

```python
import imblearn
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

```python
under = RandomUnderSampler(sampling_strategy=0.25)
new_x_training,new_y_training = under.fit_resample(x_training, y_training)
```

```python
new_y_training.value_counts()
```

```python
log=LogisticRegression()
tree=DecisionTreeClassifier()
gtree=GradientBoostingClassifier()
svm=SVC()
def classes (model_types):
   lin = model_types
   logreg=lin.fit(new_x_training,new_y_training)
   logreg.score(new_x_training,new_y_training)
   resulting=logreg.score(x_testing,y_testing)
   predict=logreg.predict(x_testing)
   matrix=confusion_matrix(y_testing,predict)
   print('the percentage accuracy is',resulting)
   print('the confusion matrix is',matrix)
classes(log)
```
